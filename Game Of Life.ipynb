```julia
SIZE = 32

F = round.(rand(SIZE, SIZE));
```

```julia
function render(F)
    for y in 1:SIZE
        for x in 1:SIZE
            if F[x, y] == 1
                print("#")
            else
                print(".")
            end
        end
        print("\n")
    end
end
```

```julia
function roll(coord)
    if coord <= 0
        return coord + SIZE
    end
    if coord > SIZE
        return coord - SIZE
    end
    return coord
end
```

```julia
function get_num_neighbours(x, y)
    count = 0
    
    for dx = -1:1
        for dy = -1:1
            if (dx != 0 || dy != 0)
                count += F[roll(x + dx), roll(y + dy)]
            end
        end
    end
    
    return count
end
```

```julia
function update_cell(x, y)
    num_neighbours = get_num_neighbours(x, y)
    
    if num_neighbours == 3
        return 1
    elseif num_neighbours == 2
        return F[x, y]
    else
        return 0
    end
end
```

```julia
function update()
    F_next = zeros(SIZE, SIZE)
    
    for y in 1:SIZE
        for x in 1:SIZE
            F_next[x, y] = update_cell(x, y)
        end
    end
    
    return F_next
end
```

```julia
for i = 1:100
    F = update();
    print("\n")
    render(F);
end
```
